```python
days = [1,2,3,4,5,6,7]
max_t = [50,51,52,48,47,49,46]
min_t = [43,42,40,44,33,35,37]
avg_t = [45,48,48,46,40,42,41]
```

```python
plt.plot(days, max_t)
plt.plot(days, min_t)
plt.plot(days, avg_t)
```

```python
plt.xlabel('Days')
plt.ylabel('temperature')
plt.title('Weather')

plt.plot(days, max_t)
plt.plot(days, min_t)
plt.plot(days, avg_t)
```

```python
plt.xlabel('Days')
plt.ylabel('temperature')
plt.title('Weather')

plt.plot(days, max_t, label = "Max")
plt.plot(days, min_t, label = 'Min')
plt.plot(days, avg_t, label = 'Avg')

plt.legend()
```

```python
plt.xlabel('Days')
plt.ylabel('temperature')
plt.title('Weather')

plt.plot(days, max_t, label = "Max")
plt.plot(days, min_t, label = 'Min')
plt.plot(days, avg_t, label = 'Avg')

plt.legend(loc='upper right')
```

```python
plt.xlabel('Days')
plt.ylabel('temperature')
plt.title('Weather')

plt.plot(days, max_t, label = "Max")
plt.plot(days, min_t, label = 'Min')
plt.plot(days, avg_t, label = 'Avg')

plt.legend(loc = 'best', shadow= True) # It will adjust automatically
```

```python
plt.xlabel('Days')
plt.ylabel('temperature')
plt.title('Weather')

plt.plot(days, max_t, label = "Max")
plt.plot(days, min_t, label = 'Min')
plt.plot(days, avg_t, label = 'Avg')

plt.legend(loc='best', shadow=True, fontsize='small')

plt.grid()
```

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
company = ['GOOGL','AMZN','MSFT','FB']
revenue = [90,136,89,27]
```

```python
ypos = np.arange(len(company))
ypos 
```

```python
plt.bar(ypos, revenue)
```

```python
plt.xticks(ypos, company) # It will replace values(ypos) to company names 

plt.bar(ypos, revenue)
```

```python
plt.xticks(ypos, company)
plt.ylabel('revenue(bln)')
plt.title('US Tech Stocks')
plt.bar(ypos, revenue)
```

```python
company = ['GOOGL','AMZN','MSFT','FB']
revenue = [90,136,89,27]
profit = [40,2,34,12]
```

```python
plt.xticks(ypos, company)
plt.ylabel('revenue(bln)')
plt.title('US Tech Stocks')
plt.bar(ypos, revenue, label='Revenue')
plt.bar(ypos, profit, label="Profit")
```

```python
plt.xticks(ypos, company)
plt.ylabel('revenue(bln)')
plt.title('US Tech Stocks')
plt.bar(ypos, revenue, label='Revenue')
plt.bar(ypos, profit, label="Profit")
plt.legend()
```

```python
plt.xticks(ypos, company)
plt.ylabel('revenue(bln)')
plt.title('US Tech Stocks')
plt.bar(ypos-0.2, revenue, label='Revenue')
plt.bar(ypos+0.2, profit, label="Profit")
plt.legend()
```

```python
plt.xticks(ypos, company)
plt.ylabel('revenue(bln)')
plt.title('US Tech Stocks')
plt.bar(ypos-0.2, revenue, width=0.4, label='Revenue')
plt.bar(ypos+0.2, profit, width=0.4, label="Profit")
plt.legend()
```

```python
plt.yticks(ypos, company)

plt.title('US Tech Stocks')
plt.barh(ypos-0.2, revenue, label='Revenue')
plt.barh(ypos+0.2, profit, label="Profit")
plt.legend()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129] # Histogram contain only single dimensions
plt.hist(blood_sugar)
```

```python
plt.hist(blood_sugar, bins=3)
```

```python
plt.hist(blood_sugar, bins=3, rwidth=0.95) # 'rwidth' means relative width
```

```python
plt.hist(blood_sugar, bins=[80,100,125,150], rwidth=0.95) # 'rwidth' means relative width
```

```python
plt.hist(blood_sugar, bins=[80,100,125,150], rwidth=0.95, color='g') 
```

```python
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100 ]

plt.hist([blood_sugar_men, blood_sugar_women], bins=[80,100,125,150], rwidth=0.95, color=['green','orange'])
```

```python
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100 ]

plt.xlabel('sugar range')
plt.ylabel('Total no of patients')
plt.title('Blood Sugar analysis')
plt.hist([blood_sugar_men, blood_sugar_women], bins=[80,100,125,150], rwidth=0.95, color=['green','orange'], label=['men','women'])
plt.legend()
```

```python
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100 ]

plt.xlabel('sugar range')
plt.ylabel('Total no of patients')
plt.title('Blood Sugar analysis')
plt.hist([blood_sugar_men, blood_sugar_women], bins=[80,100,125,150], rwidth=0.95, color=['green','orange'],
         label=['men','women'], orientation='horizontal')
plt.legend()
```
