```python
days = [1,2,3,4,5,6,7]
max_t = [50,51,52,48,47,49,46]
min_t = [43,42,40,44,33,35,37]
avg_t = [45,48,48,46,40,42,41]
```

```python
plt.plot(days, max_t)
plt.plot(days, min_t)
plt.plot(days, avg_t)
```

```python
plt.xlabel('Days')
plt.ylabel('temperature')
plt.title('Weather')

plt.plot(days, max_t)
plt.plot(days, min_t)
plt.plot(days, avg_t)
```

```python
plt.xlabel('Days')
plt.ylabel('temperature')
plt.title('Weather')

plt.plot(days, max_t, label='max')
plt.plot(days, min_t, label='min')
plt.plot(days, avg_t, label='avg')
plt.legend()
```

```python
plt.xlabel('Days')
plt.ylabel('temperature')
plt.title('Weather')

plt.plot(days, max_t, label='max')
plt.plot(days, min_t, label='min')
plt.plot(days, avg_t, label='avg')
plt.legend(loc='upper right')
```

```python
plt.xlabel('Days')
plt.ylabel('temperature')
plt.title('Weather')

plt.plot(days, max_t, label='max')
plt.plot(days, min_t, label='min')
plt.plot(days, avg_t, label='avg')
plt.legend(loc='best', shadow='Ture')
```

```python
plt.xlabel('Days')
plt.ylabel('temperature')
plt.title('Weather')

plt.plot(days, max_t, label='max')
plt.plot(days, min_t, label='min')
plt.plot(days, avg_t, label='avg')
plt.legend(loc='best', shadow='True',fontsize='small')
plt.grid()
```

```python
import matplotlib.pyplot as plt
```

```python
exp_vals = [1400, 600, 300, 410, 250]
exp_labels = ["Home Rent","Food", "phone/Internet Bill", "car", "other utilities"]
```

```python
plt.pie(exp_vals, labels=exp_labels)
```

```python
plt.pie(exp_vals, labels=exp_labels)
plt.axis("equal")
```

```python
# if you get ride of all this things we can use "plt.show()"
plt.axis("equal")
plt.pie(exp_vals, labels=exp_labels)
plt.show()
```

```python
# if you get ride of all this things we can use "plt.show()"
plt.axis("equal")
plt.pie(exp_vals, labels=exp_labels, radius=1.5)
plt.show()
```

```python
# if you get ride of all this things we can use "plt.show()"
plt.axis("equal")
plt.pie(exp_vals, labels=exp_labels, autopct='%0.1f%%', shadow=True) #autopct gives decimal values.It is a formate string
plt.show()
```

```python

plt.axis("equal")
plt.pie(exp_vals, labels=exp_labels, radius=1.5, autopct='%0.1f%%', shadow=True)
plt.show()
```

```python

plt.axis("equal")
plt.pie(exp_vals, labels=exp_labels, radius=1.5, autopct='%0.1f%%', shadow=True, explode=[0,0.1,0.1,0,0])
plt.show()
```

```python

plt.axis("equal")
plt.pie(exp_vals, labels=exp_labels, radius=1.5, autopct='%0.1f%%', shadow=True, explode=[0,0.1,0.1,0,0], startangle=180)
plt.show()
```

```python

plt.axis("equal")
plt.pie(exp_vals, labels=exp_labels, radius=1.5, autopct='%0.1f%%', shadow=True, explode=[0,0.1,0.1,0,0], startangle=90)
plt.show()
```
